```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
file_name = "car-copy.jpg"
img = cv2.imread(file_name)
cv2_imshow(img)
```

```python
print(f"Height : {img.shape[0]} pixels")
print(f"Width : {img.shape[1]} pixels")
print(f"Channels : {img.shape[2]} ch")
print()
#   
(height, width) = img.shape[:2]

center = (width // 2, height // 2)

print(width, height, center)
```

#Image Move

```python
move = np.float32([[1, 0, 100],
                   [0, 1, 100]])

moved = cv2.warpAffine(img, move, (width, height))

cv2_imshow(moved)
```

# Image Rotate
`cv2.getRotationMatrix2D(center, , )`

```python
rotate = cv2.getRotationMatrix2D(center, 90, 1.0)
rotate
```

```python
rotated = cv2.warpAffine(img, rotate, (width, height))
cv2_imshow(rotated)
```

#Image Resize

```python
SIZE = 200

ratio = SIZE / width #    
dimention = (SIZE, int(height * ratio)) # (  ,   )
```

` interpolation`    resize  , 

```python
resized = cv2.resize(img, dimention, interpolation=cv2.INTER_AREA)
cv2_imshow(resized)
```

```python
SIZE_BIG = 500

ratio_big = SIZE_BIG/ width
dimention_big = (SIZE_BIG, int(height * ratio_big))

resized_big =cv2.resize(img, dimention_big, interpolation= cv2.INTER_AREA)

# resized_big = cv2.resize(img, dimention_big, interpolation = cv2.INTER_AREA)
cv2_imshow(resized_big)
```
