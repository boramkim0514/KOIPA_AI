## 방법 1. list로만 만들기(권장 x)
    컬럼의 이름을 따로 지정해야 할 일 생김(귀찮)
    #시리즈는 1차원 배열, 시리즈가 모인 데이터 프로엠은 2차원 배열 


````python
sample_data = [["소민호", 34, 180.1],
                ["김땡떙", 33, 175.1],
                ["박땡땡", 28, 183.2]]

pd.DataFrame(sample_data)
````

````python
df1 = pd.DataFrame(sample_data)
df1
````

````python
df1.columns = ["이름", "나이", "키"]
df1
````

# 방법 2.dict로 만들기 -ver1
* 자주 사용하느 방법 중 하나
* 웹 페이지 스크래핑 한 후 자주 사용
```
딕셔너리(dict)의 key - 컬럼의 이름 (컬럼명)
딕셔너리(dict)의 value - "반드시" 리스트로 이루어져 있어야 한다.(시리즈)
모든 value의 리스트는 길이가 똑같아야 한다
```



````python
sample_data = {
    "이름":["소민호", "김땡땡", "박땡땡"], #data갯수3
    "나이":[34, 33, 28, 22], # data갯수 4
    "키": [180.1, 175.5, 183.2]} #data 갯수3
````

````python
df2 = pd.DataFrame(sample_data)
df2
````
