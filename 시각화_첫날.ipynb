```python
a1 = np.array([1, 2, 3, 4])
a2 = np.array([1, 4, 9, 16])
a3 = ['a', 'b', 'c', 'd']

plt.figure(figsize=(4, 4))
plt.plot(a3, a2)
plt.show()
```

```python
plt.plot(a3, a2, 'bp:', markersize=8 )
plt.title('my picture')
plt.legend('korea')
plt.show()
```

```python

x = np.arange(1, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
x, y1, y2, y3
```

```python
plt.plot(x, y1, 'rp:', markersize=8 )
plt.title('my picture')
plt.legend('korea')
plt.show()
```

```python


plt.plot(x, y1, 'gp-', markersize=8 )
plt.plot(x, y2, 'b^:', markersize=8 )
plt.plot(x, y3, 'r-.', markersize=8 )

plt.title('my picture')
plt.legend('korea')
plt.show()

```

```python
plt.plot(x, y1, 'gp-', markersize=8 )
plt.plot(x, y2, 'b*:', markersize=8 )
plt.plot(x, y3, 'r^', markersize=8 )

plt.title('my picture')
plt.grid(whcih='major', axis='both', color='cyan', alpha=0.5)
plt.legend(['green', 'blue', 'red'])
plt.show()
```

```python
plt.plot(x, y1, x, y2, x, y3, 'r^', markersize=8 )
plt.title('my picture')
plt.legend(['green', 'blue', 'red'])
plt.show()
```

```python
plt.plot(x, y1, 'gp-', markersize=8 )
plt.plot(x, y2, 'b*:', markersize=8 )
plt.plot(x, y3, 'r^', markersize=8 )

plt.title('my picture')
plt.grid(which='major', axis='both', color='green', alpha=0.2)
plt.legend(['green', 'blue', 'red'])
plt.show()
```

```python
x1 = ['A', 'B', 'C']
y1 = [80, 40, 20]
plt.subplot(121)
plt.title('좋아요')
plt.xlabel('xx')
plt.ylabel('yy')

plt.bar(x1, y1)


plt.subplot(122)
plt.xlabel('xx')
plt.ylabel('yy')
plt.title('유튜브')
plt.barh(x1, y1)
```

```python
x = np.arange(1, 21) 
y = x ** 4

plt.subplot(121)
plt.plot(x, y, 'g-', x, y, 'mo')
plt.title(' subplot - 1 ')

plt.subplot(122)
plt.plot(x, x ** 3, 'r-', x, x ** 3, 'g^')
plt.text(10, 5000, 'Good')
plt.title(' subplot - 2 ')

plt.suptitle(' Subplots ')
plt.show()
```

```python
color = ['green', 'yellow', 'blue', 'cyan']
hobby = ['여행', '독서', '자전거', '게임']

numbers = [20, 5, 7, 18]

plt.figure(figsize=(7,7))

#plt.pie(numbers, labels=hobby, colors=color, autopct='%.2f%%')
plt.pie(numbers, labels=hobby,  explode= (0.1, 0.1, 0.1, 0.1), colors=color, autopct='%.1f%%')

#plt.axis('equal')
plt.title('취미 비율')

plt.show()
```

```python
np.random.seed(100)
d = np.random.randint(1, 100, (3, 6))
x = np.arange(6)

for i, c in zip(range(len(d)), list('gby')) :
    plt.bar(x + i*0.2, d[i], width=0.2)

plt.title('상반기 과일 판매량')
plt.legend(labels=['사과', '배', '망고'])  
plt.xticks(x+0.2, labels=['1월', '2월', '3월', '4월', '5월', '6월'])
plt.yticks(np.arange(0, 100, 10))  

plt.show()  
```
