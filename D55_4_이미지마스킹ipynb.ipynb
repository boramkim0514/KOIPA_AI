#Load Image

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
file_name = 'car1.jpg'
img = cv2.imread(file_name)
cv2_imshow(img)
```

# 이미지 마스킹

```python
print(f"Height : {img.shape[0]} pixels")
print(f"Width : {img.shape[1]} pixels")
print(f"Channels : {img.shape[2]} ch")
```

```python
# 중심 좌표 구하기
(height, width) = img.shape[:2]
center = (width // 2, height // 2)

print("center coord : {}".format(center))
```

```python
mask = np.zeros(img.shape[:2], dtype=np.uint8)
cv2_imshow(mask)
```

```python
mask_circle = cv2.circle(mask, center, int(height/2), (255, 255, 255), -1)
cv2_imshow(mask)
```

```python
# 여러 비트 연산 중 and(&) 연산을 활용하면?
masked_bitwise = cv2.bitwise_and(img, img, mask=mask)
cv2_imshow(masked_bitwise)
```
