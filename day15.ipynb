```python
age = pd.cut(t['age'], [0, 18, 40, 80]) #연령대별 생존율
t.pivot_table('survived', ['sex', age], 'class')
```

```python
age = pd.qcut(t['fare'], 3) #3등급 요금별 생존율
t.pivot_table('survived', ['who', age], 'class')
```

```python
age = pd.qcut(t['fare'], 5) # 5등급 요금별 생존율
t.pivot_table('survived', ['who', age], 'class')
```

```python
age = pd.qcut(t['fare'], 10) # 5등급 요금별 생존율
t.pivot_table('survived', ['who', age], 'class')
```
